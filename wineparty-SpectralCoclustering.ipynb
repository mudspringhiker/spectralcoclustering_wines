```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("wineparty.csv")
```

```python
data
```

```python
data.shape
```

```python
wines = data["Unnamed: 1"].unique()
sensors = list(data.iloc[:, 3:].columns)
print(wines)
print(sensors)
```

```python
data_sensor_wine = data.iloc[:, 1:]
del data_sensor_wine["Unnamed: 2"]
data_sensor_wine = data_sensor_wine.set_index("Unnamed: 1")
```

```python
means = {}
for sensor in sensors:
    mean_sensor = []
    for wine in wines:
        mean_sensor.append(data_sensor_wine.loc[wine, sensor].mean())
    means[sensor] = mean_sensor
```

```python
means
```

```python
data_means = pd.DataFrame(means)
data_means
```

# Sensor Correlations

```python
sensor_correlations = pd.DataFrame.corr(data_means)
```

```python
ax = plt.figure(figsize=(15,10))
plt.pcolor(sensor_correlations)
plt.colorbar()
ax = plt.gca()
ax.set_yticks(np.arange(len(sensor_correlations.columns))+0.5)
ax.set_xticks(np.arange(len(sensor_correlations.index))+0.5)
ax.set_xticklabels(sensor_correlations.index, rotation="vertical")
ax.set_yticklabels(sensor_correlations.columns);
```

# Wine correlations

```python
wine_correlations = pd.DataFrame.corr(data_means.transpose())
```

```python
wine_correlations
```

```python
ax = plt.figure(figsize=(15,10))
plt.pcolor(wine_correlations)
plt.colorbar()
ax = plt.gca()
ax.set_yticks(np.arange(len(wine_correlations.columns))+0.5)
ax.set_xticks(np.arange(len(wine_correlations.index))+0.5)
ax.set_xticklabels(wines, rotation="vertical")
ax.set_yticklabels(wines);
```

# Clustering Wines

Wines are cabernet sauvignon (1), merlot (1), pinot noir (1), shiraz (1), zinfandel(5), beaoujolais (1) for a total of 6 types. So we're choosing n_clusters = 6.

```python
from sklearn.cluster.bicluster import SpectralCoclustering
model = SpectralCoclustering(n_clusters=6, random_state=42)
model.fit(wine_correlations)
```

```python
np.sum(model.rows_, axis=1)
```

```python
model.row_labels_
```

### Replotting the correlation based on clusters

```python
data_means_wines_clusters = data_means
```

```python
data_means_wines_clusters["Wines"] = pd.Series(wines)
```

```python
data_means_wines_clusters["Groups"] = pd.Series(model.row_labels_, index=data_means_wines_clusters.index)
```

```python
data_means_wines_clusters
```

```python
data_means_wines_clusters = data_means_wines_clusters.loc[np.argsort(model.row_labels_)]
```

```python
data_means_wines_clusters
```

```python
data_means_wines_clusters = data_means_wines_clusters.reset_index(drop=True)
```

```python
data_means_wines_clusters
```

```python
data_means_wines_clusters.info()
```

```python
correlations = pd.DataFrame.corr(data_means_wines_clusters.iloc[:, :27].transpose())
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.pcolor(wine_correlations)
plt.title('Original')
plt.axis("tight")
plt.subplot(122)
plt.pcolor(correlations)
plt.title("Rearranged")
plt.axis("tight")
```

```python
data_means_wines_clusters_T = data_means_wines_clusters.transpose()
data_means_wines_clusters_T
```

```python
data_means_wines_clusters_T.loc["Wines"]
```

```python
ax = plt.figure(figsize=(15,10))
plt.pcolor(correlations, cmap="Accent")
plt.colorbar()
ax = plt.gca()
ax.set_yticks(np.arange(len(correlations.columns))+0.5)
ax.set_xticks(np.arange(len(correlations.index))+0.5)
ax.set_xticklabels(data_means_wines_clusters_T.loc["Wines"], rotation="vertical")
ax.set_yticklabels(data_means_wines_clusters_T.loc["Wines"]);
```
